```python
# import the required software libraries in the order they are used in the notebook
import matplotlib.pyplot as plt
import numpy as np
# Statistics.
import scipy.stats as stats
```

```python
# define variables - no of values to be generated, mean and standard deviation
n, mu, sigma = 100000, 10.0, 3.0
```

```python
# define random generator
rng = np.random.default_rng()

# randomly generate the requested values
values = rng.normal(mu, sigma, size=n)
```

```python
# create a display object
fig, ax = plt.subplots()

# plot histogram of the data
ax.hist(values, bins=30)
ax.set_title(f'Random Distribution of {n} Values, Mean = {mu}, SD = {sigma}')
ax.set_xlabel('Value')
ax.set_ylabel('Count');
```

```python
max = values.max()
min = values.min()
```

```python
x_axis = np.linspace(min, max,n)
```

```python
# https://numpy.org/doc/2.0/reference/random/generated/numpy.random.normal.html
def normal_dist(x , mean , sd):
    prob_density = 1/(sd*np.sqrt(2 * np.pi))*np.exp(-(x-mean)**2 /(2 * sd**2))
    return prob_density
```

```python
pdf=normal_dist(x_axis, mu, sigma)
```

```python
stats.norm.pdf(values)
```

```python
fig, ax = plt.subplots()

# Plot the normal distribution.
plt.plot(x_axis, stats.norm.pdf(x_axis, mu, sigma))
plt.hist(values, density=True, bins=100)

plt.show()
```

```python
# https://numpy.org/doc/2.0/reference/random/generated/numpy.random.normal.html
plt.plot(x_axis, 1/(sigma * np.sqrt(2 * np.pi)) *

               np.exp( - (x_axis - mu)**2 / (2 * sigma**2) ),

         linewidth=2, color='r')
plt.hist(values, bins='auto', density=True)

plt.show()
```
