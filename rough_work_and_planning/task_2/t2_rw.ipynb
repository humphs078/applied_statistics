```python
# import the required software libraries in the order they are used in the notebook
import matplotlib.pyplot as plt
import numpy as np
# Statistics.
import scipy.stats as stats
```

```python
# define variables - no of values to be generated, mean and standard deviation
n, mu, sigma = 100000, 10.0, 3.0
```

```python
# define random generator
rng = np.random.default_rng()

# randomly generate the requested values
values = rng.normal(mu, sigma, size=n)
```

```python
# create a display object
fig, ax = plt.subplots()

# plot histogram of the data
ax.hist(values, bins=30)
ax.set_title(f'Random Distribution of {n} Values, Mean = {mu}, SD = {sigma}')
ax.set_xlabel('Value')
ax.set_ylabel('Count');
```

```python
# https://builtin.com/artificial-intelligence/gaussian-naive-bayes
def normal_dist(x , mean , sd):
      prob_density = (1/sd*np.sqrt(2*np.pi)) * np.exp(-0.5*((x-mean)/sd)**2)
      return prob_density
```

```python
max = values.max()
min = values.min()
```

```python
x_axis = np.linspace(min, max,n)
```

```python
pdf=normal_dist(x_axis, mu, sigma)
```

```python
# adapted from code found here - https://stackoverflow.com/questions/5484922/secondary-axis-with-twinx-how-to-add-to-legend Accessed 21 Dec. 2023

# create an empty plot
fig, ax = plt.subplots(figsize=(18,6))

g1 = ax.plot(x_axis, pdf, color = 'red', label="PDF (Manual Calculation)")
ax2 = ax.twinx()
g2 = ax2.hist(values, bins=1000, label="Randomly Generated Values")
ax.set_title('Histogram & Probability Density Function for Randomly Generated Normal Distribution')

# display legends
ax.legend(loc=1)
ax2.legend(loc=2)

ax.set_xlabel("Data Points")
ax.set_ylabel("Probability")
ax2.set_ylabel("Counts")

plt.tight_layout()
plt.show()
```
